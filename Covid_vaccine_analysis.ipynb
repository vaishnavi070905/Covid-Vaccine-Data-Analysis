## (g) Visualization

###  Top 10 States - Total Vaccination

```python
df_sorted = df.sort_values(by='Total_Doses', ascending=False).head(10)

plt.figure(figsize=(8,6))
sns.barplot(data=df_sorted, y="State", x="Total_Doses")

plt.title("Top 10 States - Total Vaccination")
plt.xlabel("Total Number of Vaccinations")
plt.ylabel("State")

plt.show()
```

### First Dose vs Second Dose

```python
# Pie chart for total doses
first_total = df['First_Dose'].sum()
second_total = df['Second_Dose'].sum()

labels = ['First Dose', 'Second Dose']
values = [first_total, second_total]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("First Dose vs Second Dose Vaccination")
plt.show()
```
